# Feature Engineering

````python
df.head()
````

## 1. Year

- So lets convert year to more readable ***age***

````python
from datetime import datetime
````

````python
current_year = datetime.now().year
age = current_year - df['year']
plt.scatter(age, df.selling_price)
````

**Higher selling price for newer cars !!**

````python
df['age'] = current_year - df['year']  # add age column
````

````python
# Drop the 'year' column (not needed anymore)
df = df.drop('year', axis=1)
````

````python
df.head(10)
````

````python
# ------------------ SEATS MERGING ------------------
def merge_seats(x):
    if 2 <= x <= 4:
        return '2-4'
    elif x > 5:
        return '>5'
    else:
        return '5'   # Only when seat = 5


````

````python
# Removing rows where mileage = 0 because a car cannot have zero mileage.
# These are incorrect/missing entries and can mislead the analysis.
df = df.loc[df.mileage != 0].copy()

# Cap selling_price at the 95th percentile to reduce extreme outliers 
# so the model is not influenced by unusually high values
outlier_th = df.selling_price.quantile(0.95)
df.loc[df.selling_price > outlier_th, 'selling_price'] = outlier_th
````

````python
 df['seats'] = df['seats'].apply(merge_seats)
````

## 2. Categorical Encoding

````python
one_hot_cols = ['seller_type', 'fuel_type', 'transmission_type', 'seats']

for col in one_hot_cols:
    dummies = pd.get_dummies(df[col], prefix=col, drop_first = True).astype(int)
    df = pd.concat([df, dummies], axis=1)

df.drop(columns=one_hot_cols, inplace=True)
````

#### 2.3 make (too many categoies)

````python
df.make.nunique()
````

````python
df['make'] = df.groupby('make')['selling_price'].transform('median')
df['model'] = df.groupby('model')['selling_price'].transform('median')
````

````python
df.head()
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df = df_scaled
````

## Evaluation Metric

#### What is R2 Score, or the coefficient of determination?

#### Let's check the performance of our previously trained model on test data.

\
LinearRegression's default `.score()` function uses R2 score to evaluate the data.

````python
model.score(X_train, y_train)
````

````python
model.score(X_test, y_test)
````

````python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# -------------------- PREDICTIONS --------------------
y_pred_train = model.predict(X_train)
y_pred_test  = model.predict(X_test)

# -------------------- METRICS ------------------------
# R Score
r2_train = r2_score(y_train, y_pred_train)
r2_test  = r2_score(y_test, y_pred_test)

# MAE
mae = mean_absolute_error(y_test, y_pred_test)

# MSE
mse = mean_squared_error(y_test, y_pred_test)

# RMSE
rmse = np.sqrt(mse)

# -------------------- PRINT RESULTS -------------------
print("R (Train):", r2_train)
print("R (Test):", r2_test)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
````
